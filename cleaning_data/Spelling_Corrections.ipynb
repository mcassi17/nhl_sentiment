```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from nltk import word_tokenize
import enchant
import re
import preprocessor as p
```

```python
col_dict = {"team": lambda x: x.strip("[]").replace('\'', '').split(", ")}
hockey = pd.read_csv('/Users/matthewcassi/Documents/nhl_sentiment/team_dfs/no_hash_url_mentions.csv', converters=col_dict)
```

```python
hockey = hockey.dropna()
```

```python
# row wise function to remove URLs and emojis. takes a row and a column
def remove_urls(row, col):
    # set tweet preprocessing to remove emojis and URLs
    p.set_options(p.OPT.URL, p.OPT.EMOJI, p.OPT.HASHTAG, p.OPT.MENTION, p.OPT.SMILEY)
    
    # remove emojis and URLs from the column
    if type(row[col]) == str:
        return p.clean(row[col])
```

```python
hockey['tweet_1'] = hockey.apply(remove_urls, axis=1, col='tweet_1')
```

```python
d = enchant.Dict("en_US")
non_dict_words = list(set([word.encode('ascii', 'ignore') for word in word_tokenize(' '.join(hockey['tweet_1'])) if d.check(word) is False and re.match('^[a-zA-Z ]*$',word)] ))
non_dict_words
```

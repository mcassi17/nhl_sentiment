```python
tweets = ' '.join(hockey['stop_text'])
```

```python
tokens = nltk.tokenize.word_tokenize(tweets)
freq = FreqDist(tokens)
```

```python
freq1 = freq.most_common(21)
```

```python
word, count = zip(*freq1)
```

```python
word = list(word)
count = list(count)
word.remove(word[0])
count.remove(count[0])
```

```python
plt.figure(figsize=(15,10))
plt.bar(np.arange(len(word)), count)
plt.xticks(np.arange(len(word)), word, rotation=45, size='large')
plt.ylabel('Frequency')
plt.title('Most Frequent Words')
plt.savefig('freq_words.png')
plt.show()
```

```python
tweets2 = tweets.replace('@', '')
tokens2 = nltk.tokenize.word_tokenize(tweets2)
bigrams = nltk.bigrams(tokens2)
bigram_freq = FreqDist(bigrams)
```

```python
freq2 = bigram_freq.most_common(21)
```

```python
bigram, bi_count = zip(*freq2)
bigram = list(bigram)
bi_count = list(bi_count)
bigram.remove(bigram[0])
bi_count.remove(bi_count[0])
```

```python
plt.figure(figsize=(15,10))
plt.bar(np.arange(len(bigram)), bi_count)
plt.xticks(np.arange(len(bigram)), bigram, rotation=65, size='large')
plt.ylabel('Frequency')
plt.title('Most Frequent Bigrams')
plt.savefig('bigrams.png')
plt.show()
```
